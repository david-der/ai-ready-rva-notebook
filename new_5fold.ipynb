```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import os

# Create directory for saving images if it doesn't exist
os.makedirs('./img', exist_ok=True)

# Set plotting style
plt.style.use('ggplot')
sns.set(style="whitegrid")
sns.set_context("notebook", font_scale=1.2)

# Load the data
print("Loading the UCI Credit Card dataset...")
df = pd.read_csv('./data/UCI_Credit_Card.csv')

# Display basic information
print("Data shape:", df.shape)
print("\nFirst few rows:")
display(df.head())

# Rename the target variable for convenience
df = df.rename(columns={'default.payment.next.month': 'default'})

# Check class distribution
print("\nClass distribution (default rate):")
default_counts = df['default'].value_counts()
default_percentages = df['default'].value_counts(normalize=True) * 100
print(f"Class 0 (No Default): {default_counts[0]} samples ({default_percentages[0]:.2f}%)")
print(f"Class 1 (Default): {default_counts[1]} samples ({default_percentages[1]:.2f}%)")
print(f"Imbalance ratio: 1:{default_counts[0]/default_counts[1]:.2f}")

# Add a note about the class imbalance
print("\nNOTE: This dataset has class imbalance which may affect model performance.")
print("We'll use balanced class weights to handle this issue.")

# Separate features and target variable
X = df.drop(['ID', 'default'], axis=1)
y = df['default']

# ======================= 5-FOLD CROSS-VALIDATION =======================
print("\n===== LOGISTIC REGRESSION WITH 5-FOLD CROSS-VALIDATION =====")

# We'll use stratified K-fold to maintain the class distribution in each fold
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize lists to store performance metrics for each fold
fold_accuracy = []
fold_precision = []
fold_recall = []
fold_f1 = []
fold_auc = []

# Initialize figure for ROC curves
roc_fig = plt.figure(figsize=(10, 8))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Chance')

# DataFrame to store feature importances across folds
feature_importance_df = pd.DataFrame(index=X.columns)

# Loop through each fold
for fold, (train_index, test_index) in enumerate(skf.split(X, y)):
    # Split data for this fold
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    print(f"\n----- Fold {fold+1}/{n_splits} -----")
    print(f"Training samples: {len(X_train)}")
    print(f"Testing samples: {len(X_test)}")
    
    # Train logistic regression model with class balancing
    logreg = LogisticRegression(
        max_iter=5000, 
        random_state=42, 
        solver='liblinear', 
        class_weight='balanced'  # Add class weighting to handle imbalance
    )
    logreg.fit(X_train, y_train)
    
    # Make predictions
    y_pred = logreg.predict(X_test)
    y_pred_prob = logreg.predict_proba(X_test)[:, 1]
    
    # Calculate metrics with zero_division parameter to handle cases with no predictions
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    
    # Store metrics
    fold_accuracy.append(accuracy)
    fold_precision.append(precision)
    fold_recall.append(recall)
    fold_f1.append(f1)
    
    # Print metrics for this fold
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    
    # Generate confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['No Default', 'Default'],
                yticklabels=['No Default', 'Default'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix - Fold {fold+1}')
    plt.savefig(f'./img/logreg_cm_fold_{fold+1}.png', bbox_inches='tight', dpi=300)
    plt.close()
    
    # Calculate and plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    fold_auc.append(roc_auc)
    
    # Use the roc_fig reference to ensure we're adding to the correct figure
    plt.figure(roc_fig.number)
    plt.plot(fpr, tpr, lw=2, alpha=0.7,
             label=f'Fold {fold+1} (AUC = {roc_auc:.3f})')
    
    # Store feature coefficients
    feature_importance_df[f'Fold_{fold+1}'] = logreg.coef_[0]
    
    # Display classification report for this fold
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

# ======================= CROSS-VALIDATION RESULTS =======================
print("\n===== CROSS-VALIDATION RESULTS =====")

# Calculate average metrics across folds
avg_accuracy = np.mean(fold_accuracy)
avg_precision = np.mean(fold_precision)
avg_recall = np.mean(fold_recall)
avg_f1 = np.mean(fold_f1)
avg_auc = np.mean(fold_auc)

# Calculate standard deviation of metrics across folds
std_accuracy = np.std(fold_accuracy)
std_precision = np.std(fold_precision)
std_recall = np.std(fold_recall)
std_f1 = np.std(fold_f1)
std_auc = np.std(fold_auc)

# Create a summary DataFrame
summary_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'],
    'Mean': [avg_accuracy, avg_precision, avg_recall, avg_f1, avg_auc],
    'Std Dev': [std_accuracy, std_precision, std_recall, std_f1, std_auc],
    'Min': [min(fold_accuracy), min(fold_precision), min(fold_recall), min(fold_f1), min(fold_auc)],
    'Max': [max(fold_accuracy), max(fold_precision), max(fold_recall), max(fold_f1), max(fold_auc)]
})

print("Summary statistics across all folds:")
display(summary_df)

# Visualize the cross-validation metrics
metrics_df = pd.DataFrame({
    'Fold': np.repeat(range(1, n_splits+1), 5),
    'Metric': np.tile(['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'], n_splits),
    'Value': np.concatenate([
        np.array([fold_accuracy[i], fold_precision[i], fold_recall[i], fold_f1[i], fold_auc[i]])
        for i in range(n_splits)
    ])
})

plt.figure(figsize=(14, 8))
sns.barplot(x='Metric', y='Value', hue='Fold', data=metrics_df)
plt.title('Performance Metrics Across Folds', fontsize=16)
plt.xlabel('Metric', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.ylim(0, 1)
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.7)  # Reference line
plt.legend(title='Fold')
plt.savefig('./img/logreg_cv_metrics.png', bbox_inches='tight', dpi=300)
plt.show()

# Visualize the stability of each metric across folds
plt.figure(figsize=(14, 8))
for i, metric in enumerate(['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC']):
    plt.plot(range(1, n_splits+1), 
             metrics_df[metrics_df['Metric'] == metric]['Value'].values,
             marker='o', label=metric, linewidth=2)
plt.title('Stability of Metrics Across Folds', fontsize=16)
plt.xlabel('Fold', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.ylim(0, 1)
plt.xticks(range(1, n_splits+1))
plt.grid(True, alpha=0.3)
plt.legend()
plt.savefig('./img/logreg_cv_stability.png', bbox_inches='tight', dpi=300)
plt.show()

# Complete the ROC curve plot
plt.figure(roc_fig.number)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.title('ROC Curves for Each Fold', fontsize=16)
plt.legend(loc="lower right")
plt.savefig('./img/logreg_cv_roc.png', bbox_inches='tight', dpi=300)
plt.show()

# ======================= FEATURE IMPORTANCE ANALYSIS =======================
print("\n===== FEATURE IMPORTANCE ANALYSIS =====")

# Calculate mean and standard deviation of feature importance across folds
feature_importance_df['Mean_Coefficient'] = feature_importance_df.mean(axis=1)
feature_importance_df['Std_Coefficient'] = feature_importance_df.std(axis=1)
feature_importance_df = feature_importance_df.sort_values('Mean_Coefficient', ascending=False)

# Display feature importance
print("Feature importance across folds (sorted by mean coefficient):")
display(feature_importance_df[['Mean_Coefficient', 'Std_Coefficient']])

# Plot feature importance with error bars
plt.figure(figsize=(14, 10))
# Fix the append (deprecated) method with pd.concat
top_features = pd.concat([feature_importance_df.head(15), feature_importance_df.tail(5)])
plt.errorbar(
    x=top_features['Mean_Coefficient'],
    y=top_features.index,
    xerr=top_features['Std_Coefficient'],
    fmt='o',
    capsize=5
)
plt.axvline(x=0, color='gray', linestyle='--')
plt.title('Top and Bottom Feature Importances with Standard Deviation', fontsize=16)
plt.xlabel('Coefficient Value', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('./img/logreg_cv_feature_importance.png', bbox_inches='tight', dpi=300)
plt.show()

# Plot the variation of the top 10 most important features across folds
top10_features = feature_importance_df.head(10).index
top10_importance = feature_importance_df.loc[top10_features, ['Fold_1', 'Fold_2', 'Fold_3', 'Fold_4', 'Fold_5']]
top10_importance = top10_importance.reset_index()
top10_importance = pd.melt(top10_importance, 
                           id_vars='index', 
                           value_vars=['Fold_1', 'Fold_2', 'Fold_3', 'Fold_4', 'Fold_5'],
                           var_name='Fold', 
                           value_name='Coefficient')

plt.figure(figsize=(14, 8))
sns.boxplot(x='index', y='Coefficient', data=top10_importance, palette='viridis')
plt.title('Variation of Top 10 Feature Coefficients Across Folds', fontsize=16)
plt.xlabel('Feature', fontsize=14)
plt.ylabel('Coefficient Value', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('./img/logreg_cv_top10_variation.png', bbox_inches='tight', dpi=300)
plt.show()

# ======================= CONCLUSION =======================
print("\n===== CONCLUSION =====")

print(f"The logistic regression model achieved an average accuracy of {avg_accuracy:.4f} (Â±{std_accuracy:.4f}) across 5 folds.")
print(f"Average F1 score: {avg_f1:.4f} (Â±{std_f1:.4f})")
print(f"Average AUC: {avg_auc:.4f} (Â±{std_auc:.4f})")

print("\nTop 5 most important features for predicting default:")
for feature, coef in zip(feature_importance_df.head(5).index, feature_importance_df['Mean_Coefficient'].head(5)):
    print(f"- {feature}: {coef:.4f}")

print("\nModel stability assessment:")
if std_accuracy < 0.02 and std_f1 < 0.02:
    print("- The model is stable across folds with low variance in performance metrics.")
elif std_accuracy < 0.05 and std_f1 < 0.05:
    print("- The model shows moderate stability across folds.")
else:
    print("- The model shows high variance across folds, indicating potential instability.")

print("\nRecommendations:")
print("1. The features with the highest importance are the most recent payment history indicators.")
print("2. Using these key features could help create a more parsimonious model.")
print("3. For production use, consider implementing a calibrated probability threshold based on business requirements.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import os

# Create directory for saving images if it doesn't exist
os.makedirs('./img', exist_ok=True)

# Set plotting style
plt.style.use('ggplot')
sns.set(style="whitegrid")
sns.set_context("notebook", font_scale=1.2)

# Load the data
print("Loading the UCI Credit Card dataset...")
df = pd.read_csv('./data/UCI_Credit_Card.csv')

# Display basic information
print("Data shape:", df.shape)
print("\nFirst few rows:")
display(df.head())

# Rename the target variable for convenience
df = df.rename(columns={'default.payment.next.month': 'default'})

# Check class distribution
print("\nClass distribution (default rate):")
default_counts = df['default'].value_counts()
default_percentages = df['default'].value_counts(normalize=True) * 100
print(f"Class 0 (No Default): {default_counts[0]} samples ({default_percentages[0]:.2f}%)")
print(f"Class 1 (Default): {default_counts[1]} samples ({default_percentages[1]:.2f}%)")
print(f"Imbalance ratio: 1:{default_counts[0]/default_counts[1]:.2f}")

# ======================= RULE-BASED SYSTEM =======================
print("\n===== RULE-BASED SYSTEM =====")
print("Rule: Predict default if PAY_0 > 0 (customer was delayed in last payment)")

# Check the distribution of PAY_0 values
print("\nPAY_0 value distribution:")
pay0_counts = df['PAY_0'].value_counts().sort_index()
display(pay0_counts)

# Define the meaning of PAY_0 values
pay_values = {
    -2: 'No consumption',
    -1: 'Paid in full',
    0: 'Revolving credit',
    1: '1 month delay',
    2: '2 months delay',
    3: '3 months delay',
    4: '4 months delay',
    5: '5 months delay',
    6: '6 months delay',
    7: '7 months delay',
    8: '8 months delay',
    9: '9+ months delay'
}

# Map PAY_0 values to their meanings
pay0_with_meaning = pd.DataFrame({
    'PAY_0': pay0_counts.index,
    'Meaning': [pay_values.get(val, 'Unknown') for val in pay0_counts.index],
    'Count': pay0_counts.values
})
display(pay0_with_meaning)

# Apply the rule-based prediction
df['rule_prediction'] = (df['PAY_0'] > 0).astype(int)

# Calculate the rule-based model's performance
rule_accuracy = accuracy_score(df['default'], df['rule_prediction'])
rule_precision = precision_score(df['default'], df['rule_prediction'])
rule_recall = recall_score(df['default'], df['rule_prediction'])
rule_f1 = f1_score(df['default'], df['rule_prediction'])

print(f"\nRule-based model performance:")
print(f"Accuracy: {rule_accuracy:.4f}")
print(f"Precision: {rule_precision:.4f}")
print(f"Recall: {rule_recall:.4f}")
print(f"F1 Score: {rule_f1:.4f}")

# Generate and display confusion matrix for the rule-based model
cm_rule = confusion_matrix(df['default'], df['rule_prediction'])
plt.figure(figsize=(10, 8))
sns.heatmap(cm_rule, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Default', 'Default'],
            yticklabels=['No Default', 'Default'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Rule-Based Model')
plt.savefig('./img/rule_confusion_matrix.png', bbox_inches='tight', dpi=300)
plt.show()

# Display classification report
print("\nClassification Report for Rule-Based Model:")
print(classification_report(df['default'], df['rule_prediction']))

# ======================= LOGISTIC REGRESSION MODEL =======================
print("\n===== LOGISTIC REGRESSION MODEL (FOR COMPARISON) =====")

# Separate features and target variable
X = df.drop(['ID', 'default', 'rule_prediction'], axis=1)
y = df['default']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Train a logistic regression model
logreg = LogisticRegression(max_iter=5000, random_state=42, solver='liblinear', class_weight='balanced')
logreg.fit(X_train, y_train)

# Make predictions
y_pred_prob = logreg.predict_proba(X_test)[:, 1]

# Find optimal threshold
print("\nFinding optimal threshold for logistic regression...")
thresholds = [0.5, 0.4, 0.3, 0.2, 0.1]
best_threshold = 0.5
best_f1 = 0

for threshold in thresholds:
    # Apply threshold to predictions
    y_pred = (y_pred_prob >= threshold).astype(int)
    
    # Calculate F1 score with this threshold
    threshold_f1 = f1_score(y_test, y_pred)
    
    print(f"Threshold {threshold}: F1 = {threshold_f1:.4f}")
    
    # Track best threshold
    if threshold_f1 > best_f1:
        best_f1 = threshold_f1
        best_threshold = threshold

print(f"\nBest threshold: {best_threshold} (F1 = {best_f1:.4f})")

# Use the best threshold
y_pred = (y_pred_prob >= best_threshold).astype(int)

# Calculate metrics
logreg_accuracy = accuracy_score(y_test, y_pred)
logreg_precision = precision_score(y_test, y_pred)
logreg_recall = recall_score(y_test, y_pred)
logreg_f1 = f1_score(y_test, y_pred)

print(f"\nLogistic Regression model performance:")
print(f"Accuracy: {logreg_accuracy:.4f}")
print(f"Precision: {logreg_precision:.4f}")
print(f"Recall: {logreg_recall:.4f}")
print(f"F1 Score: {logreg_f1:.4f}")

# Generate and display confusion matrix
cm_logreg = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm_logreg, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Default', 'Default'],
            yticklabels=['No Default', 'Default'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Logistic Regression')
plt.savefig('./img/logreg_confusion_matrix_simple.png', bbox_inches='tight', dpi=300)
plt.show()

# Display classification report
print("\nClassification Report for Logistic Regression:")
print(classification_report(y_test, y_pred))

# ======================= MODEL COMPARISON =======================
print("\n===== MODEL COMPARISON =====")

# Create a DataFrame for comparison
comparison_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1 Score'],
    'Rule-Based': [rule_accuracy, rule_precision, rule_recall, rule_f1],
    'Logistic Regression': [logreg_accuracy, logreg_precision, logreg_recall, logreg_f1]
})

display(comparison_df)

# Visualize the comparison
plt.figure(figsize=(12, 8))
comparison_df_plot = pd.melt(comparison_df, id_vars=['Metric'], var_name='Model', value_name='Score')
sns.barplot(x='Metric', y='Score', hue='Model', data=comparison_df_plot)
plt.title('Model Performance Comparison')
plt.xlabel('Metric')
plt.ylabel('Score')
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.legend(title='Model')
plt.tight_layout()
plt.savefig('./img/rule_vs_logreg_comparison.png', bbox_inches='tight', dpi=300)
plt.show()

# ======================= ERROR ANALYSIS =======================
print("\n===== ERROR ANALYSIS =====")

# Add the test set predictions to a DataFrame for analysis
test_results = pd.DataFrame({
    'Actual': y_test,
    'LogReg_Pred': y_pred,
    'PAY_0': X_test['PAY_0'],
    'PAY_2': X_test['PAY_2'],
    'PAY_3': X_test['PAY_3'],
    'LIMIT_BAL': X_test['LIMIT_BAL']
})

# Apply the rule to the test set
test_results['Rule_Pred'] = (test_results['PAY_0'] > 0).astype(int)

# Identify where models agree and disagree
test_results['Both_Correct'] = ((test_results['LogReg_Pred'] == test_results['Actual']) & 
                              (test_results['Rule_Pred'] == test_results['Actual']))
test_results['Both_Wrong'] = ((test_results['LogReg_Pred'] != test_results['Actual']) & 
                             (test_results['Rule_Pred'] != test_results['Actual']))
test_results['LogReg_Only_Correct'] = ((test_results['LogReg_Pred'] == test_results['Actual']) & 
                                      (test_results['Rule_Pred'] != test_results['Actual']))
test_results['Rule_Only_Correct'] = ((test_results['LogReg_Pred'] != test_results['Actual']) & 
                                    (test_results['Rule_Pred'] == test_results['Actual']))

# Calculate statistics
agreement_stats = {
    'Both Correct': test_results['Both_Correct'].mean() * 100,
    'Both Wrong': test_results['Both_Wrong'].mean() * 100,
    'LogReg Only Correct': test_results['LogReg_Only_Correct'].mean() * 100,
    'Rule Only Correct': test_results['Rule_Only_Correct'].mean() * 100
}

print("Agreement between models:")
for key, value in agreement_stats.items():
    print(f"{key}: {value:.2f}%")

# Analyze cases where the rule fails but logistic regression succeeds
logreg_advantage = test_results[test_results['LogReg_Only_Correct']]
if len(logreg_advantage) > 0:
    print("\nAnalysis of cases where Logistic Regression is correct but the rule fails:")
    print(f"Count: {len(logreg_advantage)}")
    print("\nDistribution of actual default values:")
    display(logreg_advantage['Actual'].value_counts())
    print("\nSummary statistics of key features:")
    display(logreg_advantage[['PAY_0', 'PAY_2', 'PAY_3', 'LIMIT_BAL']].describe())

# Analyze cases where the rule works but logistic regression fails
rule_advantage = test_results[test_results['Rule_Only_Correct']]
if len(rule_advantage) > 0:
    print("\nAnalysis of cases where the rule is correct but Logistic Regression fails:")
    print(f"Count: {len(rule_advantage)}")
    print("\nDistribution of actual default values:")
    display(rule_advantage['Actual'].value_counts())
    print("\nSummary statistics of key features:")
    display(rule_advantage[['PAY_0', 'PAY_2', 'PAY_3', 'LIMIT_BAL']].describe())

# ======================= CONCLUSION =======================
print("\n===== CONCLUSION =====")

better_model = "Logistic Regression" if logreg_f1 > rule_f1 else "Rule-Based"
print(f"Based on F1 score, the {better_model} model performs better for credit default prediction.")

# Calculate relative improvement
f1_improvement = abs(logreg_f1 - rule_f1) / rule_f1 * 100
print(f"The winning model shows a {f1_improvement:.2f}% relative {'improvement' if logreg_f1 > rule_f1 else 'decrease'} in F1 score.")

print("\nKey observations:")
print("1. The simple rule-based model uses only one feature (PAY_0) but achieves competitive performance.")
print("2. The logistic regression model can capture more complex patterns by using all available features.")
print("3. The rule-based model is more interpretable and easier to implement.")
print("4. The logistic regression model can be further optimized with different thresholds and feature engineering.")

print("\nRecommendations:")
print("1. Consider the trade-off between model complexity and performance gain.")
print("2. The rule-based model could be sufficient for simple business requirements.")
print("3. For more complex decision-making, the logistic regression model may provide better performance.")
print("4. A hybrid approach could combine the strengths of both models for different customer segments.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import StandardScaler
import os
from collections import defaultdict

# Create directory for saving images if it doesn't exist
os.makedirs('./img', exist_ok=True)

# Set plotting style
plt.style.use('ggplot')
sns.set(style="whitegrid")
sns.set_context("notebook", font_scale=1.2)

# Load the data
print("Loading the UCI Credit Card dataset...")
df = pd.read_csv('./data/UCI_Credit_Card.csv')

# Display basic information
print("Data shape:", df.shape)
print("\nFirst few rows:")
display(df.head())

# Rename the target variable for convenience
df = df.rename(columns={'default.payment.next.month': 'default'})

# Check class distribution
print("\nClass distribution (default rate):")
default_counts = df['default'].value_counts()
default_percentages = df['default'].value_counts(normalize=True) * 100
print(f"Class 0 (No Default): {default_counts[0]} samples ({default_percentages[0]:.2f}%)")
print(f"Class 1 (Default): {default_counts[1]} samples ({default_percentages[1]:.2f}%)")
print(f"Imbalance ratio: 1:{default_counts[0]/default_counts[1]:.2f}")

# ======================= FEATURE ENGINEERING =======================
print("\n===== FEATURE ENGINEERING =====")

# Define column groups
payment_cols = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
bill_cols = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']
pay_amt_cols = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

# 1. Payment history related features
# Count of missed payments (payment delay)
df['missed_payments_count'] = (df[payment_cols] > 0).sum(axis=1)

# Has any missed payment
df['has_missed_payment'] = df['missed_payments_count'] > 0

# Has missed payment last 3 months
df['missed_payment_last_3mo'] = (df[['PAY_0', 'PAY_2', 'PAY_3']] > 0).any(axis=1)

# Consecutive months with missed payments
def count_consecutive_delays(row):
    count = 0
    max_consecutive = 0
    for col in payment_cols:
        if row[col] > 0:
            count += 1
        else:
            max_consecutive = max(max_consecutive, count)
            count = 0
    return max(max_consecutive, count)

df['consecutive_delays'] = df.apply(count_consecutive_delays, axis=1)

# Payment status getting worse? (recent month vs. 6 months ago)
df['payment_status_worsening'] = df['PAY_0'] > df['PAY_6']

# Maximum payment delay (higher values indicate longer delays)
df['max_payment_delay'] = df[payment_cols].max(axis=1)

# Payment status improving over the last 3 months?
df['recent_improvement'] = (df['PAY_0'] < df['PAY_3'])

# 2. Bill amount related features
# Total bill amount across all months
df['total_bill_amount'] = df[bill_cols].sum(axis=1)

# Average bill amount
df['avg_bill_amount'] = df[bill_cols].mean(axis=1)

# Bill amount trend (increasing or decreasing) - with safeguards against infinity
df['bill_trend'] = np.where(
    df['BILL_AMT6'] != 0,
    (df['BILL_AMT1'] - df['BILL_AMT6']) / df['BILL_AMT6'].abs(),
    df['BILL_AMT1'].apply(lambda x: 1 if x > 0 else 0)  # If denominator is 0, use sign of numerator
)
# Clip to reasonable range to avoid extreme values
df['bill_trend'] = df['bill_trend'].clip(-10, 10)

# Bill amount volatility
df['bill_volatility'] = df[bill_cols].std(axis=1)

# Last month bill vs. average bill ratio - with safeguards
df['last_bill_vs_avg_ratio'] = np.where(
    df['avg_bill_amount'] != 0,
    df['BILL_AMT1'] / df['avg_bill_amount'],
    df['BILL_AMT1'].apply(lambda x: 1 if x > 0 else 0)
)
df['last_bill_vs_avg_ratio'] = df['last_bill_vs_avg_ratio'].clip(-10, 10)

# 3. Payment amount related features
# Total payment amount
df['total_payment_amount'] = df[pay_amt_cols].sum(axis=1)

# Average payment amount
df['avg_payment_amount'] = df[pay_amt_cols].mean(axis=1)

# Payment amount trend - with safeguards
df['payment_trend'] = np.where(
    df['PAY_AMT6'] != 0,
    (df['PAY_AMT1'] - df['PAY_AMT6']) / df['PAY_AMT6'].abs(),
    df['PAY_AMT1'].apply(lambda x: 1 if x > 0 else 0)
)
df['payment_trend'] = df['payment_trend'].clip(-10, 10)

# Number of months with zero payment
df['zero_payment_count'] = (df[pay_amt_cols] == 0).sum(axis=1)

# Has made any zero payments in the last 3 months
df['recent_zero_payment'] = (df[['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3']] == 0).any(axis=1)

# 4. Ratio and utilization features
# Payment to bill ratio (for each month) - with safeguards
for i in range(1, 7):
    bill_col = f'BILL_AMT{i}'
    pay_col = f'PAY_AMT{i}'
    ratio_col = f'payment_ratio_{i}'
    
    # Calculate payment ratio, handling division by zero and negative bill amounts
    df[ratio_col] = np.where(
        (df[bill_col] > 0),  # Only calculate ratio for positive bills
        np.minimum(df[pay_col] / df[bill_col], 5),  # Cap at 500% payment
        np.where(
            df[bill_col] < 0,  # If bill is negative (credit balance)
            0,  # Set to 0 (no payment needed)
            np.where(
                df[pay_col] > 0,  # If bill is 0 but payment was made
                1,  # Set to 1 (full payment)
                0   # Both bill and payment are 0
            )
        )
    )

# Average payment ratio
payment_ratio_cols = [f'payment_ratio_{i}' for i in range(1, 7)]
df['avg_payment_ratio'] = df[payment_ratio_cols].mean(axis=1)

# Minimum payment ratio (worst month)
df['min_payment_ratio'] = df[payment_ratio_cols].min(axis=1)

# Last month's utilization ratio (bill amount / credit limit) - with safeguards
df['utilization_ratio'] = np.where(
    df['LIMIT_BAL'] > 0,
    np.maximum(0, df['BILL_AMT1']) / df['LIMIT_BAL'],  # Only use positive bill amounts
    0  # If no credit limit, set to 0
)
df['utilization_ratio'] = df['utilization_ratio'].clip(0, 1)  # Cap at 100% utilization

# Average utilization ratio
df['avg_utilization_ratio'] = np.where(
    df['LIMIT_BAL'] > 0,
    np.maximum(0, df['avg_bill_amount']) / df['LIMIT_BAL'],
    0
)
df['avg_utilization_ratio'] = df['avg_utilization_ratio'].clip(0, 1)

# 5. Demographic and limit interactions
# Age groups - using integer codes instead of categorical
df['age_group'] = np.where(df['AGE'] < 30, 1,
                 np.where(df['AGE'] < 40, 2,
                 np.where(df['AGE'] < 50, 3,
                 np.where(df['AGE'] < 60, 4, 5))))

# Credit limit per age (younger people with high credit limits might be riskier)
df['limit_per_age'] = df['LIMIT_BAL'] / df['AGE']

# Education-adjusted credit utilization (education level might affect responsible use)
# Handle education level of 0 or other outliers
df['edu_adjusted_utilization'] = np.where(
    (df['EDUCATION'] > 0) & (df['EDUCATION'] <= 4),  # Use only valid education levels
    df['utilization_ratio'] / df['EDUCATION'],
    df['utilization_ratio']  # If invalid education level, just use utilization
)

# 6. Additional custom features (beyond the provided ones)

# Payment to limit ratio (payment amount relative to credit limit)
df['payment_to_limit_ratio'] = np.where(
    df['LIMIT_BAL'] > 0,
    df['PAY_AMT1'] / df['LIMIT_BAL'],
    0
)
df['payment_to_limit_ratio'] = df['payment_to_limit_ratio'].clip(0, 1)

# Debt burden: Ratio of average bill to credit limit
df['debt_burden_ratio'] = np.where(
    df['LIMIT_BAL'] > 0,
    df['avg_bill_amount'] / df['LIMIT_BAL'],
    0
)
df['debt_burden_ratio'] = df['debt_burden_ratio'].clip(0, 5)

# Payment consistency: Standard deviation of payment ratios
df['payment_ratio_std'] = df[payment_ratio_cols].std(axis=1)

# Calculate delinquency score (weighted sum of payment status)
# More recent delays are weighted more heavily
weights = [6, 5, 4, 3, 2, 1]
df['delinquency_score'] = sum(df[col] * weight for col, weight in zip(payment_cols, weights)) / sum(weights)

# Last 3-month repayment trend: Improving (1), Stable (0), or Worsening (-1)
df['recent_repayment_trend'] = np.where(
    df['PAY_0'] < df['PAY_3'], 1,  # Improving
    np.where(df['PAY_0'] > df['PAY_3'], -1, 0)  # Worsening or Stable
)

# Feature for high bill amount relative to usual with low payment
df['high_bill_low_payment'] = np.where(
    (df['BILL_AMT1'] > df['avg_bill_amount'] * 1.2) &  # Bill 20% higher than average
    (df['PAY_AMT1'] < df['avg_payment_amount'] * 0.8),  # Payment 20% lower than average
    1, 0
)

# Check for infinity or NaN values in the dataframe and replace them
df = df.replace([np.inf, -np.inf], np.nan)
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(0)  # Replace NaN only in numeric columns

print("Shape after feature engineering:", df.shape)

# Display a sample of the data with new features
print("\nSample of data with engineered features:")
display(df.sample(3))

# Check correlations with target for our new features
new_features = [
    'missed_payments_count', 'has_missed_payment', 'missed_payment_last_3mo',
    'consecutive_delays', 'payment_status_worsening', 'max_payment_delay',
    'recent_improvement', 'total_bill_amount', 'avg_bill_amount', 'bill_trend',
    'bill_volatility', 'last_bill_vs_avg_ratio', 'total_payment_amount',
    'avg_payment_amount', 'payment_trend', 'zero_payment_count',
    'recent_zero_payment', 'avg_payment_ratio', 'min_payment_ratio',
    'utilization_ratio', 'avg_utilization_ratio', 'limit_per_age',
    'edu_adjusted_utilization', 'age_group', 'payment_to_limit_ratio',
    'debt_burden_ratio', 'payment_ratio_std', 'delinquency_score',
    'recent_repayment_trend', 'high_bill_low_payment'
]

# Calculate correlation with target for new features
correlations = {}
for feature in new_features:
    correlations[feature] = df[feature].corr(df['default'])

# Create DataFrame for visualization
corr_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Correlation'])
corr_df = corr_df.sort_values('Correlation', ascending=False)

# Display correlations
print("\nFeature correlations with default:")
display(corr_df)

# Plot top correlations
plt.figure(figsize=(14, 10))
sns.barplot(x='Correlation', y='Feature', data=corr_df.head(15), palette='viridis')
plt.title('Top 15 Features by Correlation with Default', fontsize=16)
plt.xlabel('Correlation Coefficient', fontsize=14)
plt.ylabel('Feature', fontsize=14)
plt.axvline(x=0, color='gray', linestyle='--')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('./img/feature_correlation.png', bbox_inches='tight', dpi=300)
plt.show()

# ======================= RANDOM FOREST MODEL =======================
print("\n===== RANDOM FOREST MODEL =====")

# Define the features to use for modeling
# We'll use the original features + our engineered features, excluding the target and ID
base_features = ['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE'] + payment_cols + bill_cols + pay_amt_cols
all_features = base_features + new_features

# Separate features and target
X = df[all_features]
y = df['default']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Training set: {X_train.shape[0]} samples")
print(f"Testing set: {X_test.shape[0]} samples")

# Train a baseline Random Forest model
print("\nTraining baseline Random Forest model...")
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=10,
    min_samples_leaf=5,
    max_features='sqrt',
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)
rf.fit(X_train, y_train)

# Make predictions
y_pred_prob = rf.predict_proba(X_test)[:, 1]

# Find optimal threshold
print("\nFinding optimal threshold for Random Forest...")
thresholds = [0.5, 0.4, 0.3, 0.2, 0.1]
best_threshold = 0.5
best_f1 = 0

for threshold in thresholds:
    # Apply threshold to predictions
    y_pred = (y_pred_prob >= threshold).astype(int)
    
    # Calculate F1 score with this threshold
    threshold_f1 = f1_score(y_test, y_pred)
    
    print(f"Threshold {threshold}: F1 = {threshold_f1:.4f}")
    
    # Track best threshold
    if threshold_f1 > best_f1:
        best_f1 = threshold_f1
        best_threshold = threshold

print(f"\nBest threshold: {best_threshold} (F1 = {best_f1:.4f})")

# Use the best threshold
y_pred = (y_pred_prob >= best_threshold).astype(int)

# Calculate metrics
rf_accuracy = accuracy_score(y_test, y_pred)
rf_precision = precision_score(y_test, y_pred)
rf_recall = recall_score(y_test, y_pred)
rf_f1 = f1_score(y_test, y_pred)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_prob)
rf_auc = auc(fpr_rf, tpr_rf)

print(f"\nRandom Forest model performance:")
print(f"Accuracy: {rf_accuracy:.4f}")
print(f"Precision: {rf_precision:.4f}")
print(f"Recall: {rf_recall:.4f}")
print(f"F1 Score: {rf_f1:.4f}")
print(f"AUC: {rf_auc:.4f}")

# Generate and display confusion matrix
cm_rf = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Default', 'Default'],
            yticklabels=['No Default', 'Default'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Random Forest')
plt.savefig('./img/rf_confusion_matrix.png', bbox_inches='tight', dpi=300)
plt.show()

# Display classification report
print("\nClassification Report for Random Forest:")
print(classification_report(y_test, y_pred))

# Generate ROC curve and calculate AUC
plt.figure(figsize=(10, 8))
plt.plot(fpr_rf, tpr_rf, color='darkorange', lw=2, 
         label=f'ROC curve (area = {rf_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.title('ROC Curve - Random Forest', fontsize=16)
plt.legend(loc="lower right", fontsize=12)
plt.grid(alpha=0.3)
plt.savefig('./img/rf_roc_curve.png', bbox_inches='tight', dpi=300)
plt.show()

# ======================= FEATURE IMPORTANCE =======================
print("\n===== FEATURE IMPORTANCE =====")

# Get feature importances from the trained model
feature_importances = rf.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
})
feature_importance_df = feature_importance_df.sort_values('Importance', ascending=False)

print("Top 20 features by importance:")
display(feature_importance_df.head(20))

# Plot feature importances
plt.figure(figsize=(14, 10))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(20), palette='viridis')
plt.title('Feature Importance - Random Forest', fontsize=16)
plt.xlabel('Importance', fontsize=14)
plt.ylabel('Feature', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('./img/rf_feature_importance.png', bbox_inches='tight', dpi=300)
plt.show()

# Group feature importance by feature type
feature_groups = defaultdict(list)

# Categorize base features
for feature in base_features:
    if feature in ['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE']:
        feature_groups['Demographics'].append(feature)
    elif feature in payment_cols:
        feature_groups['Payment Status'].append(feature)
    elif feature in bill_cols:
        feature_groups['Bill Amounts'].append(feature)
    elif feature in pay_amt_cols:
        feature_groups['Payment Amounts'].append(feature)

# Categorize engineered features
for feature in new_features:
    if any(term in feature.lower() for term in ['miss', 'delay', 'consecutive', 'status', 'improvement', 'worsen']):
        feature_groups['Payment Behavior'].append(feature)
    elif any(term in feature.lower() for term in ['bill', 'volatility']):
        feature_groups['Bill Patterns'].append(feature)
    elif any(term in feature.lower() for term in ['payment', 'pay_amt', 'zero']):
        feature_groups['Payment Patterns'].append(feature)
    elif any(term in feature.lower() for term in ['ratio', 'utilization']):
        feature_groups['Utilization & Ratios'].append(feature)
    elif any(term in feature.lower() for term in ['age', 'edu', 'demographic']):
        feature_groups['Demographic Interactions'].append(feature)
    else:
        feature_groups['Other'].append(feature)

# Calculate aggregate importance for each group
group_importance = {}
for group, features in feature_groups.items():
    group_imp = sum(
        feature_importance_df[feature_importance_df['Feature'] == feat]['Importance'].values[0]
        for feat in features
        if feat in feature_importance_df['Feature'].values
    )
    group_importance[group] = group_imp

# Create DataFrame for plotting
group_importance_df = pd.DataFrame({
    'Feature Group': list(group_importance.keys()),
    'Importance': list(group_importance.values())
})
group_importance_df = group_importance_df.sort_values('Importance', ascending=False)

# Plot group importances
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature Group', data=group_importance_df, palette='viridis')
plt.title('Feature Group Importance', fontsize=16)
plt.xlabel('Cumulative Importance', fontsize=14)
plt.ylabel('Feature Group', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('./img/feature_group_importance.png', bbox_inches='tight', dpi=300)
plt.show()

# ======================= HYPERPARAMETER TUNING =======================
print("\n===== HYPERPARAMETER TUNING =====")

# Define a smaller parameter grid for demonstration
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 15, None],
    'min_samples_split': [10, 20],
    'min_samples_leaf': [5, 10]
}

print("Performing grid search for hyperparameter tuning...")
print("This may take some time...")

# Create a grid search
grid_search = GridSearchCV(
    RandomForestClassifier(
        random_state=42,
        class_weight='balanced',
        n_jobs=-1
    ),
    param_grid=param_grid,
    cv=3,  # 3-fold cross-validation for speed
    scoring='f1',
    n_jobs=-1,
    verbose=1
)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters and score
print("\nBest parameters found:")
print(grid_search.best_params_)
print(f"Best F1 score: {grid_search.best_score_:.4f}")

# Train a final model with the best parameters
print("\nTraining final Random Forest model with best parameters...")
best_rf = RandomForestClassifier(
    **grid_search.best_params_,
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)
best_rf.fit(X_train, y_train)

# Make predictions
y_pred_prob_best = best_rf.predict_proba(X_test)[:, 1]

# Use the previously found best threshold
y_pred_best = (y_pred_prob_best >= best_threshold).astype(int)

# Calculate metrics
best_rf_accuracy = accuracy_score(y_test, y_pred_best)
best_rf_precision = precision_score(y_test, y_pred_best)
best_rf_recall = recall_score(y_test, y_pred_best)
best_rf_f1 = f1_score(y_test, y_pred_best)
fpr_best_rf, tpr_best_rf, _ = roc_curve(y_test, y_pred_prob_best)
best_rf_auc = auc(fpr_best_rf, tpr_best_rf)

print(f"\nFinal Random Forest model performance:")
print(f"Accuracy: {best_rf_accuracy:.4f}")
print(f"Precision: {best_rf_precision:.4f}")
print(f"Recall: {best_rf_recall:.4f}")
print(f"F1 Score: {best_rf_f1:.4f}")
print(f"AUC: {best_rf_auc:.4f}")

# ======================= SIMPLE RULE-BASED SYSTEM COMPARISON =======================
print("\n===== RULE-BASED SYSTEM COMPARISON =====")
print("Rule: Predict default if PAY_0 > 0 (customer was delayed in last payment)")

# Apply the rule-based prediction to the test set
X_test_with_rule = X_test.copy()
X_test_with_rule['rule_prediction'] = (X_test_with_rule['PAY_0'] > 0).astype(int)

# Calculate the rule-based model's performance
rule_accuracy = accuracy_score(y_test, X_test_with_rule['rule_prediction'])
rule_precision = precision_score(y_test, X_test_with_rule['rule_prediction'])
rule_recall = recall_score(y_test, X_test_with_rule['rule_prediction'])
rule_f1 = f1_score(y_test, X_test_with_rule['rule_prediction'])

# Generate ROC curve for the rule-based model
# For rule-based predictions, we need to convert to binary (0/1) first
y_rule_binary = X_test_with_rule['rule_prediction']
# Calculate false positive rate and true positive rate
fpr_rule, tpr_rule, _ = roc_curve(y_test, y_rule_binary)
rule_auc = auc(fpr_rule, tpr_rule)

print(f"\nRule-based model performance:")
print(f"Accuracy: {rule_accuracy:.4f}")
print(f"Precision: {rule_precision:.4f}")
print(f"Recall: {rule_recall:.4f}")
print(f"F1 Score: {rule_f1:.4f}")
print(f"AUC: {rule_auc:.4f}")

# ======================= MODEL COMPARISON =======================
print("\n===== MODEL COMPARISON =====")

# Create a DataFrame for comparison
comparison_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'],
    'Rule-Based': [rule_accuracy, rule_precision, rule_recall, rule_f1, rule_auc],
    'Random Forest (Baseline)': [rf_accuracy, rf_precision, rf_recall, rf_f1, rf_auc],
    'Random Forest (Tuned)': [best_rf_accuracy, best_rf_precision, best_rf_recall, best_rf_f1, best_rf_auc]
})

display(comparison_df)

# Visualize the comparison
plt.figure(figsize=(14, 8))
comparison_df_plot = pd.melt(comparison_df, id_vars=['Metric'], var_name='Model', value_name='Score')
sns.barplot(x='Metric', y='Score', hue='Model', data=comparison_df_plot)
plt.title('Model Performance Comparison', fontsize=16)
plt.xlabel('Metric', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.legend(title='Model')
plt.tight_layout()
plt.savefig('./img/model_comparison.png', bbox_inches='tight', dpi=300)
plt.show()

# Plot combined ROC curves for all models
plt.figure(figsize=(12, 9))
plt.plot(fpr_rule, tpr_rule, color='red', lw=2, 
         label=f'Rule-Based Model (AUC = {rule_auc:.3f})')
plt.plot(fpr_rf, tpr_rf, color='blue', lw=2, 
         label=f'Random Forest - Baseline (AUC = {rf_auc:.3f})')
plt.plot(fpr_best_rf, tpr_best_rf, color='green', lw=2, 
         label=f'Random Forest - Tuned (AUC = {best_rf_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Chance')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.title('ROC Curve Comparison', fontsize=16)
plt.legend(loc="lower right", fontsize=12)
plt.grid(alpha=0.3)
plt.savefig('./img/combined_roc_curves.png', bbox_inches='tight', dpi=300)
plt.show()

# ======================= CONCLUSION =======================
print("\n===== CONCLUSION =====")

# Find best model based on F1 score
models = ['Rule-Based', 'Random Forest (Baseline)', 'Random Forest (Tuned)']
f1_scores = [rule_f1, rf_f1, best_rf_f1]
best_model_idx = f1_scores.index(max(f1_scores))
best_model = models[best_model_idx]

print(f"Based on F1 score, the {best_model} model performs best for credit default prediction.")

# Calculate improvement over the rule-based model
rf_improvement = (best_rf_f1 - rule_f1) / rule_f1 * 100

if best_rf_f1 > rule_f1:
    print(f"The tuned Random Forest model shows a {rf_improvement:.2f}% improvement in F1 score over the simple rule-based model.")
else:
    print(f"The simple rule-based model outperforms the Random Forest model by {-rf_improvement:.2f}% in F1 score.")

print("\nKey observations:")
print("1. The most important features for prediction are:", ", ".join(feature_importance_df['Feature'].head(5).tolist()))
print(f"2. The most important feature groups are: {', '.join(group_importance_df['Feature Group'].head(3).tolist())}")

print("\nRecommendations:")
print("1. Focus on monitoring payment history patterns, especially recent payment delays.")
print("2. Consider using a hybrid approach: simple rules for clear cases, and the Random Forest model for borderline cases.")
print("3. Set different threshold values based on business requirements (e.g., lower threshold for higher recall if needed).")
print("4. Key engineered features that improved model performance include:")
for feature in feature_importance_df[feature_importance_df['Feature'].isin(new_features)].head(5)['Feature'].tolist():
    print(f"   - {feature}")
```
